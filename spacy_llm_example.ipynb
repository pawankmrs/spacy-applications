```python
import os
import glob

contents = []
json_dir_name = '/content/drive/MyDrive/broad_ner'

import pandas as pd
from glob import glob

files = glob(json_dir_name+'/*.json')
# files = [files[0]]
df = pd.concat([pd.read_json(f_name, lines=True) for f_name in files])

```

```python
df
```

```python
#df['entities'], df['tokens'],  df['annotation_offsets']
training_data = [
  ("Tokyo Tower is 333m tall.", [(0, 11, "BUILDING")]),
]
df = df.dropna(subset=['entities', 'tokens', 'annotation_offsets'])
for index, row in df.iterrows():
    #print(row['entities'], row['tokens'], row['annotation_offsets'])
    spans = []
    for e, t, a in zip(row['entities'], row['tokens'], row['annotation_offsets']):
      if e!= 'O':
        spans.append((a[0], a[1], e))
    example = (row['text'], spans)
    training_data.append(example)

```

```python
import spacy
from spacy.tokens import DocBin

nlp = spacy.blank("en")
# training_data = [
#   ("Tokyo Tower is 333m tall.", [(0, 11, "BUILDING")]),
# ]
# the DocBin will store the example documents
db = DocBin()
for text, annotations in training_data:
    doc = nlp(text)
    ents = []
    for start, end, label in annotations:
        span = doc.char_span(start, end, label=label)
        ents.append(span)
    #print(ents)
    if len(ents)>0 and None not in ents:
      doc.ents = ents
      db.add(doc)
db.to_disk("/content/drive/MyDrive/ner_training_spacy/train.spacy")
```
